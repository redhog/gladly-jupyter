```python
import gladly_jupyter as gl
import pandas as pd
import numpy as np
```

```python
t  = np.linspace(0, 10, 1000_000, dtype="float32")
df = pd.DataFrame({"time": t, "signal": np.sin(t)+0.1 * np.sin(t*1000), "noise": np.cos(t*2), })

config = {
    "layers": [{"points": {"xData": "input.time", "yData": "input.signal", "vData": "input.noise"}}],
    "axes":   {
        "xaxis_bottom": {"label": "Time (s)"},
        "yaxis_left":   {"label": "Signal"},
        "noise": {"colorscale": "viridis", "colorbar": "horizontal"}
    },
}

gl.Plot(config, df)
```
