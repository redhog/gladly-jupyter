```python
import gladly_jupyter as gl
import pandas as pd
import numpy as np
```

```python

t = np.linspace(0, 10, 100_000, dtype="float32")
df = pd.DataFrame({"time": t, "signal": np.sin(t), "signal2": np.cos(t*2)})

gl.Plot(
    df,
    layers=[{"points": {"xData": "input.time", "yData": "input.signal", "vData": "input.signal2"}}],
    axes={
        "xaxis_bottom": {"label": "Time (s)"},
        "yaxis_left":   {"label": "Signal"},
        "signal2": {"colorscale": "viridis", "colorbar": "horizontal"}
    },
)
```
